```python
demo = pandas.read_csv('acs2017_county_data.csv')
```

```python
c = ['State','County', 'TotalPop', 'Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']
demo = demo[c]
```

```python
demo['Pacific'] = demo['Pacific'].replace({0.0:1}, inplace=True)
```

```python
demo['H'] = demo['TotalPop'] * demo['Hispanic'] //100
demo['W'] = demo['TotalPop'] * demo['White'] //100
demo['B'] = demo['TotalPop'] * demo['Black'] //100
demo['N'] = demo['TotalPop'] * demo['Native'] //100
demo['A'] = demo['TotalPop'] * demo['Asian'] //100
demo['O'] = demo['TotalPop'] * demo['Pacific'] //100
```

```python
demo[demo is np.nan] = 1.0
```

```python
demo
```

```python
demo_state = demo.groupby(['State']).sum()
```

```python
demo_state
```

```python
us = demo_state.sum()
us
```

```python
c2 = ['TotalPop', 'H', 'W', 'B', 'N', 'A', 'O']
demo_state = demo_state[c2]
```

```python
demo_state = demo_state.drop(demo_state.index[-13])
```

```python
s2 = demo_state[['TotalPop']]
```
